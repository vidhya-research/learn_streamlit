```python
import base64

def img2str_file():
    # Encode the image to a base64 string.
    with open("bg.jpg", "rb") as img:
        sb = base64.b64encode(img.read())
        s = sb.decode('utf-8')
        s="data:image/jpg;base64,"+s
    with open("img_str.txt","w") as f:
        f.write(s)

img2str_file()
```
